```python
# 1. 匯入訓練資料
import pandas as pd
train_df = pd.read_csv("train.csv", encoding="utf-8")
train_df
```

```python
test_df = pd.read_csv("test.csv", encoding="utf-8")
test_df
```

```python
# 2. 補缺失值
# 數值類型:中位數、類別類型:最常出現值
```

```python
# 資料集各欄位的缺失值總數，找出中位數並且填補
train_df.isna().sum()
test_df.isna().sum()
```

```python
med = train_df.median()
train_df = train_df.fillna(med)
test_df = test_df.fillna(med)
```

```python
train_df.isna().sum()
```

```python
# 登船口:C：Cherbourg、Q：Queenstown、S：Southampton
most = train_df["Embarked"].value_counts().idxmax()
train_df["Embarked"] = train_df["Embarked"].fillna(most)
test_df["Embarked"] = train_df["Embarked"].fillna(most)
```

```python
train_df.isna().sum()
```

```python
# 3. One-Hot Encoding
#目的:多個類別的分類變量轉換為機器學習模型可以處理的數字形式。
#核心:每個向量，向量的長度等於所有可能的類別數量，該向量中只有一個元素為1，其他元素為0。
```

```python
def nameflow(s):
  #s = "Braund, Mr. Owen Harris"
  s = s.split(",")[-1].split(".")[0]
  return s.strip()
train_df["Name"].apply(nameflow).value_counts()
```

```python
def nameflow(s):
  reserved = ["Mr","Mrs","Miss","Master"]
  s = s.split(",")[-1].split(".")[0]
  s = s.strip()
  reserved = ["Mr","Mrs","Miss","Master"]
  if s in reserved:
    return s
  else:
    return None
```

```python
train_df["Name"] = train_df["Name"].apply(nameflow)
test_df["Name"] = test_df["Name"].apply(nameflow)
test_df["Name"]
```

```python
# one-hot encoding
train = pd.get_dummies(train_df, columns=["Name", "Sex", "Embarked"])
test = pd.get_dummies(test_df, columns=["Name", "Sex", "Embarked"])
```

```python
train.columns # 顯示欄位名稱
train = train.drop(["PassengerId", "Ticket", "Cabin"], axis=1)
test = test.drop(["Ticket", "Cabin"], axis=1)
```

```python
x_train = train.drop(["Survived"], axis=1)
y_train = train["Survived"]
x_predict = test.drop(["PassengerId"], axis=1)
pid = test["PassengerId"]
```

```python
# 4. 模型訓練
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# 使用 GridSearchCV 尋找最佳參數
clf = RandomForestClassifier()
params = {
    "n_estimators":range(19, 25),
    "max_depth":range(5,9)
}
grid_search = GridSearchCV(clf, params, cv=10, n_jobs=4)
grid_search.fit(x_train, y_train)
print("最佳參數:", grid_search.best_params_)
print("最佳分數:", grid_search.best_score_)
```

```python
# 5. 模型評估
import numpy as np
from sklearn.model_selection import cross_val_score
```

```python
# 交叉驗證評估模型性能
clf = RandomForestClassifier(n_estimators=23,
                       max_depth=8)
scores = cross_val_score(clf,
                         x_train,
                         y_train,
                         cv=10,
                         n_jobs=4)  # 使用交叉驗證評估模型性能
print("十次分數:", scores)
print("平均分數:", np.average(scores))
```

```python
# 6. 模型預測
clf = RandomForestClassifier(n_estimators=23,
                       max_depth=8)
clf.fit(x_train, y_train)
pre = clf.predict(x_predict)
pre
```

```python
# 7. 生成預測結果文件
result = pd.DataFrame({
    "PassengerId":pid,
    "Survived":pre
})
result.to_csv("titanic_rf.csv", encoding="utf-8", index=False)
result
```
